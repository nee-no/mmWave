```python
# foo-v0-7 training loss and reward (accuracy)

import csv
import numpy as np
import os
from matplotlib import pyplot as plt
%matplotlib inline

with open('foo-vo-7/results_foo-v0-7.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

loss_data = [] #temp[11]
mean_reward = [] #temp[5]
    
for j in range(0, len(data)):
    if j%2:
        temp = data[j]
        loss = temp[11].split(':')
        loss_data.append(float(loss[1]))
        
        mean = temp[5].split(':')
        mean = mean[1].split('[')
        mean_reward.append(float(mean[0]))
loss_altered = loss_data[2:len(loss_data)-1]
plt.plot(loss_data)
plt.show()
plt.plot(mean_reward)
#print(loss_data)
```

```python
# foo-v0-7 testing

import csv
import numpy as np
import os
from matplotlib import pyplot as plt
%matplotlib inline

with open('foo-vo-7/test_rwd_foo-v0-7.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

mean_reward = [] #temp[5]
acc = []

j = 0    
while j < (len(data)):
    if (j+1):
        temp = data[j+1]
        temp = temp[0].split(':')
        acc.append(float(temp[1]))
    if (j+2):
        temp = data[j+2]
        temp = temp[0].split(',')
        temp = temp[0].split(':')
        mean_reward.append(float(temp[2]))
    j += 3
                
plt.plot(mean_reward)
# accuracy = total num of correct guesses / total num of guesses
avg_reward = sum(mean_reward)/(len(mean_reward)*5)
avg = sum(acc)/(len(acc)*5)
print(avg_reward, avg)
```
